### numpy 的array操作

#### 1.导入numpy库

```python
import numpy as np
```

#### 2.建立一个一维数组 a 初始化为[4,5,6], (1)输出a 的类型（type）(2)输出a的各维度的大小（shape）(3)输出 a的第一个元素（值为4）

```python
a = np.array([4,5,6])
print("(1) 类型（type）:", type(a))
print("(2) 各维度大小（shape）:", a.shape)
print("(3) 第一个元素为", a[0])
```

#### 3.建立一个二维数组 b,初始化为 [ [4, 5, 6],[1, 2, 3]] (1)输出各维度的大小（shape）(2)输出 b(0,0)，b(0,1),b(1,1) 这三个元素（对应值分别为4,5,2）

```python
b=np.array([ [4, 5, 6],[1, 2, 3]])
print("(1) 各维度大小（shape）:", b.shape)
print("(2) b[0,0] =", b[0, 0], ", b[0,1] =", b[0, 1], ", b[1,1] =", b[1, 1])
```

#### 4.  (1)建立一个全0矩阵 a, 大小为 3x3; 类型为整型（提示: dtype = int）(2)建立一个全1矩阵b,大小为4x5;  (3)建立一个单位矩阵c ,大小为4x4; (4)生成一个随机数矩阵d,大小为 3x2.

```python
arr1 = np.zeros((3, 3), dtype=int)
print("(1) 3x3全0矩阵：\n", arr1)
arr2 = np.ones((4, 5))
print("(2) 4x5全1矩阵：\n", arr2)
arr3 = np.eye(4)
print("(3) 4x4单位矩阵：\n", arr3)
arr4 = np.random.random((3, 2))
print("(4) 3x2随机数矩阵：\n", arr4)
```

#### 24,画图，y=x*x 其中 x = np.arange(0, 100, 0.1) （提示这里用到  matplotlib.pyplot 库）

```python
x = np.arange(0, 100, 0.1)
y = x * x
import matplotlib.pyplot as plt
plt.plot(x, y, 'b-')
plt.xlabel('x')
plt.ylabel('y = x^2')
plt.grid(True)
```

#### 25.画图。画正弦函数和余弦函数， x = np.arange(0, 3 * np.pi, 0.1)(提示：这里用到 np.sin() np.cos() 函数和 matplotlib.pyplot 库)

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
import matplotlib.pyplot as plt
plt.subplot(2, 1, 2)
plt.plot(x, y_sin, 'r-', label='sin(x)')
plt.plot(x, y_cos, 'g-', label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
